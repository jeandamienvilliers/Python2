## Pandas

### Loads a dataframe from Google spreadsheet and build pivot tables

Loads the file: dividends_SX5E as a pandas dataframe

Display the 5 first rows of the dataframe

Create a pivot table which reports for each ticker the maximum price.

### Google credentials

```python
from google.colab import auth
auth.authenticate_user()

import gspread
from google.auth import default
creds, _ = default()

gc = gspread.authorize(creds)
```

### Load Excel file

```python
worksheet=gc.open('dividends_SX5E').sheet1
worksheet
```

### Convert it to a dataframe

```python
import pandas as pd
all_values=worksheet.get_all_values()
df=pd.DataFrame.from_records(all_values[1:len(all_values)])
df.columns=all_values[0]
df
```

### Display first 5 rows

```python
df.head(5)
```

### Create a pivot table with the max price for each ticker

```python
pd_max_price=pd.pivot_table(data=df, index=["Ticker"], values=["Last Price"], aggfunc=np.max)
pd_max_price
```
